Df + coluna dias ( do momento da compra ate chegar) e previsão do tempo da compra até chegar.

```python
df['order_purchase_timestamp'] = pd.to_datetime(df['order_purchase_timestamp'])
df['order_delivered_customer_date'] = pd.to_datetime(df['order_delivered_customer_date'])
df['order_estimated_delivery_date'] = pd.to_datetime(df['order_estimated_delivery_date'])

df['dias_para_chegar'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days
df['dias_para_chegar_estimado'] = (df['order_estimated_delivery_date'] - df['order_purchase_timestamp']).dt.days

display(df[['order_purchase_timestamp', 'order_delivered_customer_date', 'dias_para_chegar', 'order_estimated_delivery_date', 'dias_para_chegar_estimado']].head())
display(df['dias_para_chegar'].isnull().sum())
display(df['dias_para_chegar'].dtype)
display(df['dias_para_chegar_estimado'].isnull().sum())
display(df['dias_para_chegar_estimado'].dtype)
```

```python
df2 = df.drop(columns=['order_purchase_timestamp','order_delivered_customer_date', 'order_estimated_delivery_date'], errors='ignore')
display(df.head())
```

```python
df2 = df2.dropna(subset=['dias_para_chegar', 'dias_para_chegar_estimado'])
```

```python
df2.head()
```

```python
df2['houve_atraso'] = (df2['dias_para_chegar'] > df2['dias_para_chegar_estimado'])
display(df2.head())
```

--

```python
df2.groupby('houve_atraso')['review_score'].mean()
```

```python
# 1. Carregando a tabela de Clientes do Kaggle
df_clientes = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "olistbr/brazilian-ecommerce",
  "olist_customers_dataset.csv"
)


df_completo = pd.merge(df2, df_clientes[['customer_id', 'customer_state']], on='customer_id', how='left')
```

```python
df_completo.groupby('customer_state')['houve_atraso'].sum().sort_values(ascending=False)
```

```python
taxa_atraso_uf = df_completo.groupby('customer_state')['houve_atraso'].mean().sort_values(ascending=False) * 100

print("Taxa de atraso por Estado (%):")
print(taxa_atraso_uf.head())
```
